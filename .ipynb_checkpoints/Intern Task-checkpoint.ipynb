```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://topai.tools/browse'
```

```python
response=requests.get(url)
response
```

```python
s=response.text
s
```

```python
soup=BeautifulSoup(response.text,'html.parser')
soup
```

```python
# print(type(soup))
# soup=(str(soup))
```

```python
tags_with_class = soup.find_all(class_='mt-3 mb-2 px-1')
tags_with_class
```

```python
# print(type(tags_with_class))
```

```python
tags_with_class=str(tags_with_class)
tags_with_class
```

```python
word=tags_with_class.split('<h5 class="mt-3 mb-2 px-1">\n')

information=[]

for part in word[1:]:
    information.extend(part.split("\n<a")[:-1])

Name = ",".join(information)

print(Name)
print(type(Name))
```

```python
word=tags_with_class.split("href=")

information=[]

for part in word[1:]:
    information.extend(part.split('rel="nofollow"')[:-1])

Link = ",".join(information)

print(Link)
print(type(Link))
```

```python
word=tags_with_class.split('<span class="badge float-end bg-black mr-2 pricing-badge">')

information=[]

for part in word[1:]:
    information.extend(part.split("</span>")[:-1])

Price = ",".join(information)

print(Price)
print(type(Price))
```

```python
tags_with_class = soup.find_all(class_='my-1 font-weight-lighter px-1')
tags_with_class
```

```python
tags_with_class=str(tags_with_class)
tags_with_class
```

```python
word=tags_with_class.split('<p class="my-1 font-weight-lighter px-1" style="height:80px;font-size:15px">')

information=[]

for part in word[1:]:
    information.extend(part.split("..</p>")[:-1])

Tool_use_case = ";".join(information)

print(Tool_use_case)
print(type(Tool_use_case))
```

```python
print(Name)
print(Link)
print(Price)
print(Tool_use_case)
```
